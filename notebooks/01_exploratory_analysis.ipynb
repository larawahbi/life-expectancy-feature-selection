# 01 — Exploratory Data Analysis

Before selecting features or building models, we need to understand 
the data structure, quality, distributions, and how each variable 
relates to life expectancy.

This notebook answers four questions:
1. What does the data look like and what is missing?
2. How is life expectancy distributed across countries and development status?
3. Which variables correlate most strongly with life expectancy?
4. Are there any variables we should be cautious about before modelling?

**Input:** `../data/Life Expectancy Data.csv`  
**Output:** Cleaned dataset `../data/life_expectancy_clean.csv`

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('../data/Life Expectancy Data.csv')

# Strip whitespace from column names - the original has trailing spaces
df.columns = df.columns.str.strip()

print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nData types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum())

# Results: Missing values are significant in some columns.
# Population is missing 652 rows (22% of data), GDP 448 (15%), Hepatitis B 553 (19%)
# Population is missing so heavily and doesn't directly predict life expectancy at the country level — it's more a size variable than a health determinant. We'll examine whether to drop it. GDP and Hepatitis B are important predictively but missing for many rows — we'll impute those.
```
